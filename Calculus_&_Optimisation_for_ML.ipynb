Chain rule

```python
#approach 1
# y = (2x+3)^2

# q = 2x+3
# y = q^2

# Step 1: Differentiate y with respect to q
# dy/dq = 2q

# Step 2: Differentiate q with respect to x
# dq/dx=2

# Step 3: Apply the Chain Rule
# dy/dx= dy/dq *dq/dx
#      = 2q*2= 4q



# Step 4: Substitute back q=2x+3

# dy/dx = 4(2x+3)
#       = 8x+12




```

```python
#approach 2: using expansion
# (a+b)^2 = a^2+2ab+b^2

# y = (2x+3)^2 = 4x^2+12x+9

# dy/dx = 4x^2+12x+9
#       = 8x+12


```

Gradient Descent Algo


```python
#this is algo or method used in ML to reduce error step by step
```

```python
# the parameter we are refering is weigth vector
```

```python
# Gradient --> slope direction
# Descent --> moving downword
# loss --> error (Diff b/w the actual and predicted)
# learning rate --> step size
```

```python
# # formula
# new_weight = old_weight - learning_rate*gradient
```

```python
# initial_weight =10
# learning_rate = 0.1
# gradient = 2

# new_weight = 10 - 0.1*2
#            = 9.8(now this will be old value)
# 10--> 9.8 -> 9.9 -> 9.4 .....
```

```python
 initial_weight =10
 learning_rate = 0.1
for i in range(1,6):
   gradient = 2

   initial_weight = initial_weight - learning_rate*gradient

   print(round(initial_weight,2))


```

```python
# impact of learning rate?
# if the learning rate is very small--- will take too mauch time
# if the learning rate is very large-- it will jump left and right so will not be able to reach correct result
# so we need to be slow and stable.
```

```python
# gradient descent ==> slope tells the direction in which erroe will increase.
# model should move opposite to the slope that is why we have - sign in formula.
# gradient is + then we need to move +
# if gradient is - then we need to move -
```

```python
# regression --> -2, 3.5, 99, 102, 38.3
# classification---> 0 or 1


#******** sigmoid function: The sigmoid function converts any value into a probability between 0 and 1.
# and probability helps in decision making
# probabilityv >= 0.5--> class1
# else: probability < 0.5 ---> class 0
# this is called decision boundary

# sigmoid function: 1/(1+e^(-x)) = e^x / (1+e^x)----> alway lies between 0 to 1

# numerator < denominator

# e^x is always positive
# sigmoid(x)--> calculate
# values of x:

# -10
# -2
# 0
# 2
# 10

```

```python
import numpy as np

def sigmoid(x):
  return 1/(1+np.exp(-x))

print(sigmoid(-10))
print(sigmoid(-2))
print(sigmoid(0))
print(sigmoid(2))
print(sigmoid(10))
```

```python
# sigmoid --> 0.85--> 85% probability of this belonging to that class
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
#loan approval on the basis of income---> approve or not approve.

income = np.array([20,30,80,100]).reshape(-1,1)
# Simple summary:
# -1 = automatic row calculation
# 1 = one column
approval = np.array([0,0,1,1]) #This is your output (labels)

# 0 = Loan Not Approved
# 1 = Loan Approved

model = LogisticRegression()
#This line is creating a Logistic Regression model


model.fit(income,approval)

# You are training the model

# Income → input
# Approval → output

# Model learns pattern:

# “Higher income → more chance of approval


print(model.predict([[70]]))
# You are asking:

# “For income = 70, is loan approved or not?”


# Expected idea of result
# Since:
# 20, 30 → 0 (not approved)
# 80, 100 → 1 (approved)
#  70 is closer to 80
# So model will likely predict:

# [1]


```

# Confusion Matrix

```python
# A confusion matrix is a table that shows how well a classification model is performing.
#  It compares:
# Actual values vs Predicted values
```

```python
# ### Why Confusion Matrix came into picture (Easy Summary)

# Accuracy alone was not enough to evaluate a model properly.

# ---

# ### Problem

# * A model can show high accuracy
# * But still make serious mistakes (for example, missing sick people in medical tests)

# ---

# ### Solution

# So we use a confusion matrix to clearly show:

# * Correct predictions (TP, TN)
# * Wrong predictions (FP, FN)

# ---

# ### Simple idea

# Accuracy = one overall score
# Confusion matrix = detailed breakdown of mistakes

# ---

# ### One-line answer

# Confusion matrix was introduced to show detailed correct and incorrect predictions because accuracy alone can hide important mistakes.
```

```python
# sick/healthy
#  sick--> positive, Healthy --> negative(we are looking for sick people)
# actual---> predcited
#  sick --> healthy
#  sick --> sick
#  healthy --> healthy
#  healthy --> sick

```

```python
# 1. True Positive(TP)
# actual is sick and the model also predicted it to be sick-- 1

# 2. True Negative(TN)
# actual is healthy and a model also predicted it to be healthy-->1

# 3. False Positive(FP)
# actual is healthy and model predicted it to be sick--> 0

# 4. False Negative(FN)
# actual is sick and model predicted it to be healthy--> 0(most negative thing)

```

```python
# actuals = [1,1,1,0,0,0]
# predicted = [1,1,0,0,1,0]
#  1-> positive , 0-> negative
# TP-2
# TN-2
# FP-1
# FN-1

# confusion matrix:
```

```python
from sklearn.metrics import confusion_matrix
actual = [1,1,1,0,0,0]
predicted = [1,1,0,0,1,0]

print(confusion_matrix(actual,predicted))
```

```python
# Confusion matrix shows detailed correct and wrong predictions, while metrics like accuracy, precision, recall, and F1 score summarize model performance.
# Main Metrics” (Accuracy, Precision, Recall, F1)

# These are numbers calculated from confusion matrix

# Example:
# Accuracy = overall correct %
# Precision = correct YES predictions
# Recall = missed case check
# F1 = balance score
```

```python
# 1. Accuracy: (TP+TN)/Total
#            :(TP +TN)/(TP+TN+FP+FN)
```

```python
# Accuracy = (All correct predictions) / (All predictions)
```

```python
# 100 patients --> 99 no cancer ,1 cancer.
# model predicted all 100 to be no cancer

# model accuracy=?  99% but useless. so to over come this we have below matrix.
```

```python
# 1. Precision: out of the predictive positives, how
# many were actually positive?

# precision = TP/(TP+FP)

# 2. recall: out of the actual positivies,
# how many we also predicted to be positive?

# recall = TP/(TP+FN)

# 3. F1 Score = 2*(precision*recall)/(precision+recall)

```

```python
# Accuracy, Precision, Recall, F1 (Easy Summary)

# These are metrics used to measure how good a classification model is.
#They are all calculated using the confusion matrix (TP, TN, FP, FN).
```

```python
# Accuracy, Precision, Recall, F1

# These are NOT matrices

# They are:

# metrics (numbers) calculated from the confusion matrix
```

```python
# Accuracy = overall performance
# Precision = correctness of positive predictions

# Accuracy = checks all correct answers (positive + negative)
# Precision = checks only positive correct answers
```

```python
#
 Accuracy example

# Model correctly identifies:

# 85 normal emails
# 8 spam emails

# Total correct = 93/100
#  Accuracy = 93%

# Includes BOTH:

# correct normal emails
# correct spam emails
# Precision example (only spam predictions)

# Model says:

# 12 emails are spam
# But only 8 are actually spam

# Precision = 8/12

# Only checks “spam predictions”
```

```python
# Recall (Very Easy Explanation)

# Recall means: Out of all actual YES cases, how many did the model correctly find?

# Simple idea

# “Did I catch all the real positives?”

# Example (Disease test)

# Suppose:

# 10 people are actually sick (real YES)
# Model detects only 7 as sick

#  Recall = 7/10 = 70%
```

```python
# Memory trick
# Precision = correctness
# Recall = completeness
# F1 = balance of both
```

```python
# Accuracy checks total correctness, precision checks correctness of positive predictions, recall checks how many real positives are found,
# and F1 gives a balance of precision and recall.
```
